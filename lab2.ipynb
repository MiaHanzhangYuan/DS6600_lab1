```python

root = 'https://maps.spirithalloween.com'
domain='domain'
endpoint = f'/api/getAsyncLocations?template={domain}&level={domain}'
headers = {'User-Agent': useragent,
 'From': 'ynx9zm@virginia.edu'}
parameters = {'format': 'json',
             "render": "all",
            "limit": 2000,
            "radius": 1800,
            "zoom": "autofit",
            "mapContainerId": "gmap",
            "listContainer": ".map-list",
            "psid": 1,
            "did": "5555",
            "lat": 40.380028,
            "lng": -97.910156}
r = requests.get(root + endpoint,
                params = parameters,
                headers = headers)
```

```python
myjson = json.loads(r.text)
```

```python
result = []
for i in myjson['markers']:
    address = BeautifulSoup(i['info']).find('div', attrs={'class':'address-two'}).getText()
    result.append(address)
    
```

```python
formerlocs = [i for i in result if 'Former' in i]
```
